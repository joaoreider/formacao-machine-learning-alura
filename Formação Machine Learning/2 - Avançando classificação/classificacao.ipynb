```python
import pandas as pd
dados = pd.read_csv("dados\categoricas.csv")
dados
```

```python
x = dados[['home', 'busca', 'logado']]
y = dados['comprou']

# variáveis dummy sao variáveis categóricas especiais porque são traduzidas em binárias, dizemos se pertence a um grupo ou não. No nosso caso, a coluna busca é uma Dummy e traduzimos em 3 colunas bin
x = pd.get_dummies(x)

# separando treino e teste
from sklearn.model_selection import train_test_split
seed = 5
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, random_state = seed, test_size = 0.1, stratify=y)

# Modelo
from sklearn.naive_bayes import MultinomialNB
modelo = MultinomialNB() # cria
modelo.fit(x_treino, y_treino) # treina

# previsao
resultado = modelo.predict(x_teste)

# Taxa de acerto
from sklearn.metrics import accuracy_score
taxa_de_acerto = accuracy_score(y_teste, resultado)

print('Taxa de acerto: {}%'.format((taxa_de_acerto).round(2)*100))
print('Tamanho do teste: ', len(y_teste))
```

```python
# Como posso dizer q essa taxa de acerto é boa? Qual base?
# O algoritmo burro, prevê o mesmo numero sempre
quantidade_que_comrpou = sum(y)
tamanho_y = len(y)
base = quantidade_que_comrpou / tamanho_y
# Veja que o algortimo simples tem a mesma taxa de acerto do nosso algoritmo anterior, mas cuidado, temos que escolher a categoria que mais aparece para o algoritmo base
acerto_de_um = sum(y)
acerto_de_zero = len(y) - acerto_de_um
taxa_de_acerto_base = 100 * max(acerto_de_um, acerto_de_zero) / len(y)
print('Taxa de acerto: {}%'.format((taxa_de_acerto_base)))
```

```python
# às vezes nossas categorias não estão com 1 e 0 mas sim dois valores quaisquer, por isso existe um método útil para contar
from collections import Counter
Counter(y)
# Se fosse 'sim' ou 'nao' ia retornar com o nome da categoria, como no exemplo
# com isso podemos fazer um algoritmo de teste genérico
contador = Counter(y_teste)
acerto_base = max(contador.values()) # corrigindo que a taxa de acerto tem q ser com os dados de teste e não num geral
taxa_de_acerto_base = 100 * acerto_base / len(y_teste)
print('Taxa de acerto: {}%'.format((taxa_de_acerto_base)))
```

```python
# O modelo que utilizamos foi o MultinomialNB - Naive Bayes  -> Como é o funcionamento desse algoritmo bayiesiano?
# Também chamado de aprendizado bayesiano, em decorrência da fórmula de Bayes, é amplamente utiizado em inteligência computacional e consiste na avaliação de hipóteses
# pela máxima verossimilhança
# O treino é baseado em probabilidades condicionais 
# O teste utiliza regras de decisão
# Usado bastente para classificar textos

import pandas as pd
df = pd.read_csv('dados\web2.csv')
df
```

```python
# Utilizando outro algoritmo
from sklearn.ensemble import AdaBoostClassifier

x = df[['home', 'busca', 'logado']]
y = df['comprou']

# variáveis dummy sao variáveis categóricas especiais porque são traduzidas em binárias, dizemos se pertence a um grupo ou não. No nosso caso, a coluna busca é uma Dummy e traduzimos em 3 colunas bin
x = pd.get_dummies(x)

# separando treino e teste
from sklearn.model_selection import train_test_split
seed = 5
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, random_state = seed, test_size = 0.1, stratify=y)

# Modelo
from sklearn.naive_bayes import MultinomialNB
modelo = AdaBoostClassifier() # cria
modelo.fit(x_treino, y_treino) # treina

# previsao
resultado = modelo.predict(x_teste)

# Taxa de acerto
from sklearn.metrics import accuracy_score
taxa_de_acerto = accuracy_score(y_teste, resultado)

print('Taxa de acerto: {}%'.format((taxa_de_acerto).round(2)*100))
print('Tamanho do teste: ', len(y_teste))
```

```python
# definindo uma função que faz o fit e predict do modelo
def fit_and_predict(nome, modelo, x_treino, y_treino, x_teste, y_teste):
    modelo.fit(x_treino, y_treino) # treina
    
    resultado = modelo.predict(x_teste) # previsao

    from sklearn.metrics import accuracy_score
    taxa_de_acerto = accuracy_score(y_teste, resultado)

    print('Taxa de acerto {}: {}%'.format(nome, (taxa_de_acerto).round(2)*100))
    print('Tamanho do teste: ', len(y_teste))
```

```python
from sklearn.naive_bayes import MultinomialNB
modelo = AdaBoostClassifier() # cria
fit_and_predict('AdaBoost', modelo, x_treino, y_treino, x_teste, y_teste)

from sklearn.naive_bayes import MultinomialNB
modelo = MultinomialNB() # cria
fit_and_predict('MultinomialNB', modelo, x_treino, y_treino, x_teste, y_teste)
```
